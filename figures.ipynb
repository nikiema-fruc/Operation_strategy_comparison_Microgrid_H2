# results from sensitivity analysis

```julia
EMS1=zeros(Float64,3)
EMS2=zeros(Float64,3)
EMS1=[0.3332,0.3592,0.3871]
EMS2=[0.3587,0.4136,0.4551]

names = ["case 1", " base case  ", " case 2"]
ano = ["-7.24%", "-13.2%","+7.7%","+10%"]
x=collect(range(1,3,3))
x=2*x
fig1, ax = plt.subplots(figsize=(5*0.75,6*0.75))


st1=ax.stem(x, EMS1,"r",label="EMS1")

 for i=1:2
 ax.annotate(ano[2*i-1] ,
                   (x[2*i-1] , EMS1[2*i-1]) , ha="center", va="center",
                   size=10, xytext=(5, 10),
                   textcoords="offset points")
end


ax.stem(x .+ 0.75, EMS2,"b", label="EMS2")
for i=1:2
 ax.annotate(ano[2*i] ,
                   (x[2*i-1] + 0.75 , EMS2[2*i-1]) , ha="center", va="center",
                   size=10, xytext=(0, 10),
                   textcoords="offset points")
end

ax.set_ylim(0.3, 0.47)
ax.set_xlim(0.5,8)
ax.set_xticks(x .+ 0.375,names)
ax.set_title("LCOE for each scenario")
ax.set_ylabel("\$/kWh")
ax.legend(loc="upper left")
plt.savefig("figures/sensibility.png", dpi=300)
```
